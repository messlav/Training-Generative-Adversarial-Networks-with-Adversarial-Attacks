* Let's show some images from different iterations

```python
from PIL import Image
import glob

frames = []
imgs = glob.glob('logs/WGANGP_CIFAR10_RES/sample/*.png')
for i in imgs:
    new_frame = Image.open(i)
    frames.append(new_frame)
 
# Save into a GIF file that loops forever
frames[0].save('WGANGP_ResNet_samples.gif', format='GIF',
               append_images=frames[1:],
               save_all=True,
               duration=300, loop=0)
```
